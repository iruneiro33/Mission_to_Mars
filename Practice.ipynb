```python
from splinter import Browser
from bs4 import BeautifulSoup as soup
from webdriver_manager.chrome import ChromeDriverManager
```

```python
# Set up Splinter
executable_path = {'executable_path': ChromeDriverManager().install()}
browser = Browser('chrome', **executable_path, headless=False)
# **executable_path means unpacking the dictionary we've stored the path in
# headless= False means that all of the browser-s actions will be displayed in a Chrome window
```

```python
# Visit the Quotes to Scrape site
url = 'http://quotes.toscrape.com/'
browser.visit(url)
```

```python
# Parse the HTML
html = browser.html
html_soup = soup(html, 'html.parser')
html_soup
```

```python
# Scrape the Title
title = html_soup.find('h2').text
title
```

```python
# Scrape the top ten tags (which have "tag-item class")
tag_box = html_soup.find('div', class_='tags-box')
                        


tags = tag_box.find_all('a', class_='tag')

for tag in tags:
    word = tag.text
    print(word)
```

### Scrape Across Pages

```python
url = 'http://quotes.toscrape.com/'
browser.visit(url)
```

```python
for x in range(1,6):
    html = browser.html
    quote_soup = soup(html, 'html.parser')
    quotes = quote_soup.find_all('span', class_='text')
    for quote in quotes:
        print('page', x, '-----------') 
        print(quote.text)
    browser.links.find_by_partial_text('Next').click() #open the next page
        
        
```

```python
for x in range(1,6):
    html = browser.html
    quote_soup = soup(html, 'html.parser')
    quotes = quote_soup.find_all('div', class_='quote')
    for quote in quotes:
        print('page', x, '-----------') 
        print(quote.text)
    browser.links.find_by_partial_text('Next').click() #open the next page
        
        
```
